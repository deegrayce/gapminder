```python
#importing packages.
import pandas as pd
```

```python
#reading csv file from drive and viewing the head of the dataset
data = r"/content/gapminder_with_codes.csv"
df = pd.read_csv(data)
df.head()
```

```python
#viewing column names
df.columns
```

```python
#list of countries in the dataset
df['country'].unique()
```

```python
#unique number of countries

df['country'].nunique()
```

```python
#value counts of countries
df['country'].value_counts()
```

```python
#viewing the data of India for the year 2007
df[(df["country"] == "India") & (df["year"] == 2007)]
```

```python
#viewing the data of Nigeria for the year 2007
df[(df["country"] == "Nigeria") & (df["year"] == 2007)]
```

```python
#viewing the data of countries in Africa with gdp less than or equal to 5000 with a high population
df[(df["continent"] == "Africa") & (df["gdpPercap"] <= 5000) & (df["pop"] >= 111039600)]
```

```python
import seaborn as sns
```

```python
#importing packages
import matplotlib.pyplot as plt

#plotting gdp versus year for Nigeria
df[(df["country"] == "Nigeria")].plot(x = "year", y = "gdpPercap")
#setting the title and labels
plt.title("GDP vs Year for Nigeria")
plt.xlabel("Year")
plt.ylabel("GDP per capita")
plt.show()
```

```python
#plotting pop versus year for Nigeria
df[(df["country"] == "Nigeria")].plot(x = "year", y = "pop")

#setting the title and labels
plt.title("Population vs Year for Nigeria")
plt.xlabel("Year")
plt.ylabel("population")
plt.show()
```

```python
#plot a graph of gdp versus year for India

sns.lineplot(x="year", y="gdpPercap", data=df[df['country']=='India'])
```

```python
#plot a graph of pop versus year for India

sns.lineplot(x="year", y="pop", data=df[df['country']=='India'])
```

```python
#plot a graph of pop versus continent

sns.barplot(x="continent", y="pop", data=df)
```
